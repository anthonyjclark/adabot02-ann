```python
%matplotlib inline

# Bokeh for interactive, Seaborn for static

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

sns.set(style="whitegrid", palette="pastel")
# sns.set(style="whitegrid", palette=sns.color_palette("RdBu", n_colors=2))

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

from math import pi
import glob
```

```python
sweep_fullfile = '/Users/ajc/Documents/projects/adabot/adabot02-ann/src/extras/bnn/bin/evolved_bnn_sweep.csv'
df = pd.read_csv(sweep_fullfile)
df.head()
```

```python
# https://github.com/mGalarnyk/Python_Tutorials/blob/master/
# Request/Heat%20Maps%20using%20Matplotlib%20and%20Seaborn.ipynb

temp = df.ix[:, ['angle', 'angular error', 'left speed']]

plt.figure(figsize=(9,9))
pivot_table = temp.groupby(['angle', 'angular error']).mean().reset_index().pivot(
    'angle', 'angular error', 'left speed')
# plt.xlabel('helix 2 phase', size = 15)
# plt.ylabel('helix1 phase', size = 15)
# plt.title('Energy from Helix Phase Angles', size = 15)
sns.heatmap(pivot_table, square=True, cmap='Blues_r');
```
